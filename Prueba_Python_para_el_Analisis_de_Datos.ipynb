```python
# Primero y antes de resolver completamente el punto 1 se explora qué contienen todas las tablas
# Se genera una lista con las 8 tablas
tablas = ['customers', 'employees', 'offices', 'orderdetails', 'orders', 'payments', 'productlines', 'products']
# Se leen todas las tablas y se llevan a DataFrames a través de un diccionario para almacenar los DataFrames
dataframes ={}
for tabla in tablas:
    query = f"SELECT * FROM {tabla};"
    df = pd.read_sql(query, engine)
    dataframes[tabla] = df
# Buena práctica visualización de las primeras 5 filas
for tabla, df in dataframes.items():
    print(f"Primeras 5 filas de la tabla '{tabla}':")
    display(df.head())
```

```python
# Se procede a resolver lo solicitado en el punto 1
# Se define la función que lee una tabla y la devuelve como un DataFrame
def leer_tabla(tabla, engine):
    """
    La función lle una tabla desde la base de datos y la convierte en un DataFrame.
    
    Args:
    - tabla (str): el nombre de la tabla a leer
    - engine: el motor de conexión a la base de datos

    Returns:
    - df (DataFrame): un DataFrame con los datos de la tabla
    """
    query = f"SELECT * FROM {tabla};"
    df = pd.read_sql(query, engine)
    return df
# Se genera una lista con las 5 tablas: orders, orderdetails, customers, products, employees
tablas_punto1 = ['customers', 'employees', 'orderdetails', 'orders' , 'products']
# Se crea un diccionario para alamacenar los DataFrames
dataframes_punto1 ={}
# Se lee cada tabla y se almacena en el diccionario
for t in tablas_punto1:
    dataframes_punto1[t] = leer_tabla(t, engine)
# Buena práctica visualización de las primeras 5 filas
for t, df in dataframes_punto1.items():
    print(f"Primeras 5 filas de la tabla_punto1 '{t}':")
    display(df.head())
```

```python
# Filtrar las ventas del año 2005
df_basal['orderDate'] = pd.to_datetime(df_basal['orderDate'])  # Asegurar que la columna es tipo datetime
ventas_2005 = df_basal[df_basal['orderDate'].dt.year == 2005]

# Reporte 1: Top 10 clientes con mayores ventas brutas en 2005
top_10_clientes_2005 = (
    ventas_2005.groupby(['customerName'], as_index=False)
    .agg({
        'venta': 'sum',
        'costo': 'sum',
        'ganancia': 'sum'
    })
    .sort_values(by='venta', ascending=False)
    .head(10)
)

# Guardar el DataFrame en PostgreSQL
top_10_clientes_2005.to_sql('top_10_clientes_2005', engine, index=False, if_exists='replace')

# Mostrar el resultado
print("Top 10 clientes con mayores ventas en 2005:")
display(top_10_clientes_2005)

```

```python
# Reporte 2: Top 10 productos más vendidos por cantidad neta en 2005
top_10_productos_2005 = (
    ventas_2005.groupby(['productName'], as_index=False)
    .agg({
        'quantityOrdered': 'sum',
        'venta': 'sum',
        'costo': 'sum',
        'ganancia': 'sum'
    })
    .sort_values(by='quantityOrdered', ascending=False)
    .head(10)
)

# Guardar el DataFrame en PostgreSQL
top_10_productos_2005.to_sql('top_10_productos_2005', engine, index=False, if_exists='replace')

# Mostrar el resultado
print("Top 10 productos más vendidos por cantidad neta en 2005:")
display(top_10_productos_2005)
```
